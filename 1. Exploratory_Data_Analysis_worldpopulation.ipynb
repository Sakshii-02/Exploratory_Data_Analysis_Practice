```python
# Filter numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix: Comparing the columns and looking how closely correlated they are
correlation_matrix = numeric_df.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot = True) # To visualise the correlation
plt.rcParams['figure.figsize'] = (18,7)
plt.show()
# We can see that 2022 population is highly correlated. Also density, rank, growth is not that much correlated with the population.
```
